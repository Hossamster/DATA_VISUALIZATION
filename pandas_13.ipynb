# Bar Plots

```python
plants = ['Spanish','Turnip','Rhubarb','Broccoli','Kale']
died = [10,25,5,30,21]
germinated = [74,88,56,69,59]
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants,germinated,color = 'green',label = 'Died',width = 0.6)
plt.title('Germination rates')
plt.ylabel('Number of Seeldings')
plt.legend(loc = 'upper right')
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants,died,width = 0.5)
plt.ylabel('Number of Dead Seedlings',labelpad = 10,color ='black',fontsize = 15)
plt.title('Failure Rates')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants,died,width = 0.5,color = '#ff7675')
plt.bar(plants,germinated,width = 0.4,color = '#00cec9',align ='edge')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants,died,color = '#ff7675',bottom = germinated ,label = 'Failed Seedlings')
plt.bar(plants,germinated,color = '#00cec9',label = 'Germinated Seedlings')
plt.legend()
plt.title('Greenhouse Seedlings')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.barh(plants,died,color = '#ff7675')
```

# Subplots

```python
nums = np.arange(5)
```

```python
plt.figure(figsize =(10,4))
plt.suptitle('Comparison',fontsize = 30)
plt.subplot(1,3,1)
plt.title("X")
plt.plot(nums,nums)
plt.subplot(1,3,2)
plt.title('X squared')
plt.plot(nums,nums**2,color = 'red')
plt.subplot(1,3,3)
plt.title('X cubed')
plt.plot(nums,nums**3,color = 'green')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize =(4,10))
plt.suptitle('Comparison',fontsize = 30)

plt.subplot(3,1,1)
plt.title("X")
plt.plot(nums,nums)

plt.subplot(3,1,2)
plt.title('X squared')
plt.plot(nums,nums**2,color = 'red')

plt.subplot(3,1,3)
plt.title('X cubed')
plt.plot(nums,nums**3,color = 'green')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.suptitle('2 * 2 ')
plt.subplot(2,2,1)
plt.title('X')
plt.plot(nums,nums,)

plt.subplot(2,2,2)
plt.title('X squared')
plt.plot(nums,nums**2,)

plt.subplot(2,2,3)
plt.title('X squared')
plt.plot(nums,nums**3,)

plt.subplot(2,2,4)
plt.title('X squared')
plt.plot(nums,nums**4,)
plt.tight_layout()
```

```python
first_class = titanic[titanic['pclass'] == 1]['age']
second_class = titanic[titanic['pclass'] == 2]['age']
third_class = titanic[titanic['pclass'] == 3]['age']
plt.figure(figsize = (10,4))
plt.suptitle('Titanic Passenger ages')

ax = plt.subplot(1,3,1)
plt.title('1st class')
plt.hist(first_class,label = 'first class',alpha = 0.8,bins = 20,color = 'blue')
plt.legend()

plt.subplot(1,3,2,sharey = ax)
plt.title('2nd class')
plt.hist(second_class,label = 'second class',alpha = 0.8,bins = 20,color = 'red')
plt.legend()

plt.subplot(1,3,3,sharey = ax)
plt.title('3rd class')
plt.hist(third_class,label = 'third class',alpha = 0.8,bins = 20,color = 'green')
plt.legend()

plt.tight_layout()
plt.show()  
```
