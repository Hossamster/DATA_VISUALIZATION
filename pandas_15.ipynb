```python
def range(s):
    return s.max() - s.min()
titanic.groupby('pclass')['age'].agg(range)
```

```python
def range(s):
    return s.max() - s.min()
titanic.groupby('pclass')['age'].agg(['min','max',range])
```

```python
# How many null values
titanic['age'].size - titanic['age'].count()
```

```python
def count_nulls(s):
    return s.size - s.count()
```

```python
titanic.groupby('pclass')['age'].agg(count_nulls)
```
