## Part 2

**Use `agg` to create a dataframe that contains:**
* A `total` column containing the total "Shots" taken by each team
* A `on_target` column containing the total "Shots on target" taken by each team
* It should look like the following dataframe (but for all teams in the dataset):

| Team   | total | on_target |
| ----------- | ----------- | ----------- |
| **Real Betis** | 300      |158
| **Levante UD** | 314      |157

```python
laliga.columns
```

```python
mydf = laliga.groupby('Team').agg(total = ('Shots','sum'),on_target = ('Shots on target','sum'))
```

```python
mydf
```

```python
mydf['total'].sort_values(ascending=False)
```

```python
mydf['on_target'].sort_values(ascending = False)
```

```python
new = mydf['on_target'] / mydf['total']
```

```python
most = new.sort_values(ascending= False).head(5)
```

```python
least = new.sort_values(ascending= False).tail(5)
```

Use the dataframe from above to create the following figure:

* Notice the layout (2 rows by 1 column)
* The top chart shows the top 5 most accurate teams (highest on-target shot pecentage)
* The bottom chart shows the 5 least accurate teams (lowest on-target shot pecentage)
* Both plots share the same x-axis
* Notice how the data is sorted within each plot!




```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
most
```

```python
fig,axs = plt.subplots(2,1,figsize = (10,8))
fig.tight_layout()

p = most.plot(kind = 'barh',sharex = True,title = 'Least Accurate Teams',fontsize =20,color = 'green',ax = axs[0])
p.invert_yaxis()
axs[0].set_title('Most Accurate Teams',fontsize = 20)
axs[0].set_ylabel('Team',fontsize = 20)

k = least.plot(kind = 'barh',sharex = True,color = 'red',ax = axs[1],fontsize = 20)
axs[1].set_title('Least Accurate Teams',fontsize = 20)
axs[1].set_ylabel('Team',fontsize = 20)
k.invert_yaxis()

plt.xlabel('On Target Percentage',labelpad = 15,fontsize = 15)
plt.tight_layout()
plt.show()
```
