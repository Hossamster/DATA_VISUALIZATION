Answer the following questions using the `bestsellers.csv` dataset

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Hossam\Desktop\data visualiztion\DataAnalysisCourseMaterials\DataAnalysis\data\bestsellers.csv")
df
```

```python
#Find the books written by Pete Souza
df[df['Author'] == 'Pete Souza']['Name']
```

```python
# Find the books that are under 10 dollars
df[df['Price'] < 10]['Name']
```

```python
#Find the books that have a price between 50 and 60 dollars
df[df['Price'].between(50,60)].Name
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens
df[(df['Author'].isin(['Kristin Hannah','Andy Weir','Delia Owens']))]
```

```python
#Find the Non Fiction books that are rated 4.9
df[(df['Genre'] == 'Non Fiction') & (df['User Rating'] == 4.9)]
```

```python
#Find the fiction book with the lowest User Rating
small = df['User Rating'].sort_values()
small
df[(df['Genre'] == 'Fiction')][['Name','User Rating']].sort_values(['User Rating']).head(1)
```

```python
#Find 2012's top 5 Fiction books with the most Reviews
df[(df['Year'] == 2012) & (df['Genre']=='Fiction')].sort_values(['Reviews'],ascending = False).head(5)
```

```python
#Create a bar plot showing the 5 authors who have the most books with a rating under 4.5
df[df['User Rating'] < 4.5][['Author']].value_counts().head(5).plot(kind ='bar')
```
